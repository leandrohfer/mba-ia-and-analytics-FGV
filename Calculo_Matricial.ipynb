```python
import numpy as np

# Declaração de uma matriz
A = np.array([
    [2,3],
    [1,7]
])

# Método para obter a inversa de uma matriz
A_inv = np.linalg.inv(A)

# Multiplicando a matriz A pela sua inversa para obter a matriz Identidade
print(A_inv @ A)
```
